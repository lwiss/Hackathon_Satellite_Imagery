```python
import numpy as np 
import os
from os.path import basename
from PIL import Image 
```

```python
# get training and test file names 
tr_label_dir = '/Users/taaalwi1/Desktop/hackathon/map/train/'
tr_samples_dir = '/Users/taaalwi1/Desktop/hackathon/sat/train/'
ts_label_dir = '/Users/taaalwi1/Desktop/hackathon/map/test/'
ts_samples_dir = '/Users/taaalwi1/Desktop/hackathon/sat/test/'

tr_label_dir_new = '/Users/taaalwi1/Desktop/hackathon/map_new/train/'
tr_samples_dir_new = '/Users/taaalwi1/Desktop/hackathon/sat_new/train/'
ts_label_dir_new = '/Users/taaalwi1/Desktop/hackathon/map_new/test/'
ts_samples_dir_new = '/Users/taaalwi1/Desktop/hackathon/sat_new/test/'

samples_extention = '.tiff'
label_extention = '.tif'
## get file names 
tr_file_names_list = []
ts_file_names_list = []
for file in os.listdir(tr_label_dir):
    if file.endswith('.tif'): 
        tr_file_names_list.append(os.path.splitext( basename(file))[0])
        
for file in os.listdir(ts_label_dir):
    if file.endswith('.tif'): 
        ts_file_names_list.append(os.path.splitext( basename(file))[0])
```

```python
original_size = 1500
new_size = 224
number_of_new_images = 50

# iterate of training and generate new data both for labels and for sat images
for name in tr_file_names_list:
    # generate random corner for the new images 
    upper = np.random.randint(0, original_size-new_size, number_of_new_images)
    left_ = np.random.randint(0, original_size-new_size, number_of_new_images)
    for i, (up, left) in enumerate(zip(upper, left_)):
        print ((i,up, left))
        # crop the sat image 
        image_path = os.path.join(tr_samples_dir, name + samples_extention)
        img = Image.open(image_path)
        new_img = img.crop((up, left, up+new_size, left+new_size))
        new_img.save(os.path.join(tr_samples_dir_new ,"{0}_{1}{2}".format(name,i,samples_extention)))
        # crop the label mask 
        image_path = os.path.join(tr_label_dir, name + label_extention)
        img = Image.open(image_path)
        new_img = img.crop((up, left, up+new_size, left+new_size))
        new_img.save(os.path.join(tr_label_dir_new ,"{0}_{1}{2}".format(name,i,label_extention)))
        
# Now do the same for the test set
for name in ts_file_names_list:
    # generate random corner for the new images 
    upper = np.random.randint(0, original_size-new_size, number_of_new_images)
    left_ = np.random.randint(0, original_size-new_size, number_of_new_images)
    for i, (up, left) in enumerate(zip(upper, left_)):
        print ((i,up, left))
        # crop the sat image 
        image_path = os.path.join(ts_samples_dir, name + samples_extention)
        img = Image.open(image_path)
        new_img = img.crop((up, left, up+new_size, left+new_size))
        new_img.save(os.path.join(ts_samples_dir_new ,"{0}_{1}{2}".format(name,i,samples_extention)))
        # crop the label mask 
        image_path = os.path.join(ts_label_dir, name + label_extention)
        img = Image.open(image_path)
        new_img = img.crop((up, left, up+new_size, left+new_size))
        new_img.save(os.path.join(ts_label_dir_new ,"{0}_{1}{2}".format(name,i,label_extention)))
    
    
```
